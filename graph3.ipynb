```python
def yield__():
    a = df[(df['Year'] == 2014)]
    
    b = []
    for i in (a['State_Name']).iteritems():
        b.append(str(i[1]))
        
    b = list(set(b))
    
    n = len(b)

    for i in range(n):
        for j in range(0, n-i-1):
            x = a[(a['State_Name'] == b[j])]
            y = a[(a['State_Name'] == b[j+1])]
            if float((x['Production']).mean()) > float((y['Production']).mean()):
                b[j], b[j+1] = b[j+1], b[j]
    b.reverse()
    
    c = []
    for i, j in zip(b,range(5)):
            c.append(i)
    b = c
    print(b)
    
    li = []
    liv = []
    
    for i in b:
        d = a[(a['State_Name'] == i)]
        e = []
        for j in (d['Crop']).iteritems():
            e.append(str(j[1]))
        e = list(set(e))
        n = len(e)

        for j in range(n):
            for k in range(0, n-j-1):
                x = d[(d['Crop'] == e[k])]
                y = d[(d['Crop'] == e[k+1])]
                if float((x['Production']).mean()) > float((y['Production']).mean()):
                    e[k], e[k+1] = e[k+1], e[k]
        e.reverse()
        
        c = []
        for j, k in zip(e,range(3)):
                c.append(j)
        e = c
        li.append(e)
        o = []
        for j in e:
            x = d[(d['Crop'] == j)]
            x = float((x['Production']).mean())
            o.append(x)
        liv.append(o)
        
    print(li)
    print(liv)
yield__()
```
